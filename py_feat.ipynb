```python
import os
import random
import pandas as pd

root_folder = "/Users/khushnumaparveen/Desktop/DAiSEE/DataSet/Train"

video_paths = []
for dirpath, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.endswith((".mp4", ".avi")):
            video_paths.append(os.path.join(dirpath, filename))

sample_videos = random.sample(video_paths, 5000)

all_summaries = []
landmarks_list = []
action_units_list = []

for i, video_path in enumerate(sample_videos, 1):
    result = detector.detect_video(video_path, skip_frames=5)
    print(f"{i} Processed  - {video_path}")
    print(result.head())

    # --- Save landmarks + pitch, yaw, roll columns ---
    landmarks_cols = [col for col in result.columns if col.startswith('x_') or col.startswith('y_')]
    pose_cols = ['Pitch', 'Yaw', 'Roll']
    pose_cols = [col for col in pose_cols if col in result.columns]  # Only add if present
    landmarks_pose = result[landmarks_cols + pose_cols].copy()
    landmarks_pose['ClipID'] = os.path.basename(video_path)
    landmarks_list.append(landmarks_pose)

    # --- Save action units columns ---
    action_units_cols = [col for col in result.columns if col.startswith('AU')]
    action_units = result[action_units_cols].copy()
    action_units['ClipID'] = os.path.basename(video_path)
    action_units_list.append(action_units)

    # --- Aggregate summary for each video (mean of numeric columns) ---
    summary = result.mean(numeric_only=True)
    summary['ClipID'] = os.path.basename(video_path)
    summary['num_frames'] = len(result)
    all_summaries.append(summary)

# Save summaries (one row per video)
combined_summaries = pd.DataFrame(all_summaries)
combined_summaries.to_csv('combined_results_one_row_per_video_final.csv', index=False)

# Save all landmarks + pose data (all frames)
pd.concat(landmarks_list, ignore_index=True).to_csv('landmarks_and_pose_final.csv', index=False)

# Save all action units (all frames)
pd.concat(action_units_list, ignore_index=True).to_csv('action_units_final.csv', index=False)
```
